 <font color="black" style="font-family: Courier" > The elucidation of protein functions through their three-dimensional structures is a cornerstone of post-genomic research. As we delve deeper into the genomic era, the imperative to infer protein functionality from amino acid sequences has become ever more pressing. Understanding the complexity of protein structures can be formidable due to their atomistic intricacy. An effective strategy to manage this complexity is to adopt a hierarchical approach to protein structure. This hierarchy simplifies the protein structure into four classical levels: primary, secondary, tertiary, and quaternary structures. The primary structure denotes the linear sequence of amino acids, while the secondary structure refers to local motifs such as α-helices and β-sheets. Tertiary structure encompasses the three-dimensional conformation of a single polypeptide chain, and quaternary structure describes the assembly of multiple such chains into a functional unit.
<br>
Beyond these, supersecondary structures, or motifs, represent combinations of secondary elements that form recognizable patterns, and domains, which are distinct functional and structural units within a polypeptide. These additional levels bridge the gap between the secondary and tertiary structures and are critical in understanding the folding and functional regions of proteins.
<br>
Recent advancements have introduced additional layers to this hierarchy, such as the consideration of dynamic conformational changes (conformational ensembles) and the impact of post-translational modifications on structure and function. Such nuances add depth to our understanding of proteins beyond static structures and allow for a more nuanced appreciation of protein functionality in the cellular context.</font>

## <font color="black"> 4. Computational Protein Structure Prediction: </font>

### <font color="black"> Homology or Comparative Modeling: 
Comparative modeling, also known as homology modeling, is predicated on the notion that structural conservation is more stringently preserved than the sequence during evolution. This method infers the three-dimensional (3D) structure of a target protein based on its alignment with one or more template proteins of known structure. The process involves four primary stages: template identification, sequence alignment, model building, and model assessment. Comparative modeling is most reliable when the target and template share a sequence identity above 30%, making it a powerful tool for predicting protein structures with known homologues.
    
- Tools: MODELLER, SWISS-MODEL, Phyre2

- Example Practice: You have the sequence of a protein that's similar to hemoglobin. Using MODELLER, you'd input your sequence and select a known hemoglobin structure as your template. MODELLER then aligns your sequence to the template and constructs a model of your protein's 3D structure.

### Fold Recognition or Threading Methods:
When sequence similarity to known structures is low, fold recognition or threading becomes a vital tool. This approach evaluates the compatibility of a target sequence with established 3D structures, employing statistical parameters or potentials of mean force to gauge the quality of fit. Threading can reveal structural similarities that sequence comparisons might miss, thus enabling the identification of distant homologues and the prediction of protein folds with no apparent sequence identity to known structures.

- Tools: I-TASSER, HHpred, PHYRE

- Example Practice: Imagine your protein is like a puzzle piece with a unique shape, but you're not sure where it fits. Using I-TASSER, you can compare your piece against a database of solved puzzle shapes to find the best match and predict your protein's structure.

</font>




### Ab initio Methods Utilizing Knowledge-Based Information:
Ab initio methods aim to predict protein structure solely based on the amino acid sequence, without relying on template structures. Despite ongoing research, ab initio prediction remains a challenge, generally limited to small proteins. The use of accurate force fields to simulate atomic interactions is crucial. While physics-based force fields provide detailed interaction descriptions, they are computationally intensive. Knowledge-based potentials derived from the PDB offer a more feasible alternative, requiring careful selection of reference states and structural features to guide the folding simulation effectively.

Examples of Knowledge-Based Ab Initio Methods:

- ROSETTA:

Rosetta is one of the most well-known ab initio prediction tools that uses knowledge-based energy functions to predict protein structure.
Example Exercise: As a practice, you could take a small protein or peptide sequence and use Rosetta to predict its structure. Begin by creating a 'fragment library' that Rosetta will use to assemble a model of your protein. Then, run the 'ab initio' protocol, which will generate thousands of possible structures. Finally, select the lowest energy models and refine them to get your predicted structure.

- QUARK:

QUARK is another tool for ab initio protein structure prediction. It assembles protein fragments derived from known structures to build models for the target sequence.
Example Exercise: Similar to Rosetta, you could use QUARK for a protein of interest. Input your sequence into QUARK, which will then create a pool of fragment assemblies. QUARK will provide you with a set of possible structures ranked by their energy scores, from which you can choose potential candidates for your protein's structure.

- Knowledge-Based Information:

Potential of mean force: This is a type of knowledge-based potential derived from statistical analysis of distances and contacts between different types of amino acids in known protein structures.
Secondary structure propensity: Some sequences have a higher likelihood of forming certain secondary structures like alpha helices or beta strands. Knowledge about these propensities helps guide the folding simulation.
Solvent accessibility: The degree to which amino acids are exposed to solvent in known structures can inform the model about which parts of the sequence are likely to be on the surface or buried.

- Tutorial Exercise:

  Beginner Level: Use an online server that runs ROSETTA or QUARK to submit a sequence and interpret the results. This would involve learning how to look at the different energy components and understand what they mean for the predicted structure.<br> <br>
  Advanced Level: Install and run the ROSETTA or QUARK software locally. Start with a simple peptide to understand the process and gradually move to larger proteins. Analyze the output structures using a visualization tool like PyMOL, and compare the predicted structures to any available experimental data.

### Ab initio Methods Without Knowledge-Based Information:
<font> Ab initio Methods Without Knowledge-Based Information:

In bioinformatics, ab initio methods that do not utilize knowledge-based information rely on pure computational algorithms to predict protein structures from amino acid sequences. These methods use principles of physics and chemistry to simulate the process of protein folding from first principles. The challenge with these methods is the sheer computational power required to accurately model all the atomic interactions over the time scales necessary for a protein to fold.

<br>


**AlphaFold:**

AlphaFold2 represents a paradigm shift in ab initio protein structure prediction. Developed by DeepMind, AlphaFold2 uses a deep learning approach to predict the 3D structure of proteins with remarkable accuracy. It does not directly rely on knowledge-based databases of existing structures; instead, it learns from the Protein Data Bank indirectly through training. AlphaFold2 uses an attention-based neural network architecture to determine the distances and angles between amino acid residues in a protein, allowing it to infer the protein's likely 3D structure.
<br>
AlphaFold2's success at the 14th Critical Assessment of protein Structure Prediction (CASP14) competition has been a significant milestone, highlighting its ability to predict protein structures with high precision, sometimes rivaling experimental methods.

To predict the structure of a protein using homology modeling, you can use the Modeller software, which is a popular tool for this purpose. Modeller automates the process of creating a model based on the sequence alignment and structural templates. Below is a Python script that demonstrates how to use Modeller for homology modeling. This script is for educational purposes and assumes that you have Modeller installed in your environment.

First, ensure you have Modeller installed, which you can obtain from https://salilab.org/modeller/. Please follow the installation guide on their website to set it up correctly.

Here's a basic Python script to perform homology modeling using Modeller. Note that you'll need to know the template structure's PDB code and have an alignment file ready for this to work. For this example, I'll assume we're using a hypothetical template with PDB code 'XXXX' and an alignment file named 'alignment.ali'.

```python
from modeller import *

log.verbose()
env = Environ()

#-- Prepare the input files
import os

# Set the working directory to a specific path
os.chdir('MODELER/basic-example/')

#-- Read in the sequence database
sdb = SequenceDB(env)
sdb.read(seq_database_file='pdb_95.pir', seq_database_format='PIR',
         chains_list='ALL', minmax_db_seq_len=(30, 4000), clean_sequences=True)

#-- Write the sequence database in binary form
sdb.write(seq_database_file='pdb_95.bin', seq_database_format='BINARY',
          chains_list='ALL')

#-- Now, read in the binary database
sdb.read(seq_database_file='pdb_95.bin', seq_database_format='BINARY',
         chains_list='ALL')

#-- Read in the target sequence/alignment
aln = Alignment(env)
aln.append(file='TvLDH.ali', alignment_format='PIR', align_codes='ALL')

#-- Convert the input sequence/alignment into
#   profile format
prf = aln.to_profile()

#-- Scan sequence database to pick up homologous sequences
prf.build(sdb, matrix_offset=-450, rr_file='${LIB}/blosum62.sim.mat',
          gap_penalties_1d=(-500, -50), n_prof_iterations=1,
          check_profile=False, max_aln_evalue=0.01)

#-- Write out the profile in text format
prf.write(file='build_profile.prf', profile_format='TEXT')

#-- Convert the profile back to alignment format
aln = prf.to_alignment()

#-- Write out the alignment file
aln.write(file='build_profile.ali', alignment_format='PIR')
```

```python
from modeller import *

env = Environ()
aln = Alignment(env)
for (pdb, chain) in (('1b8p', 'A'), ('1bdm', 'A'), ('1civ', 'A'),
                     ('5mdh', 'A'), ('7mdh', 'A'), ('1smk', 'A')):
    m = Model(env, file=pdb, model_segment=('FIRST:'+chain, 'LAST:'+chain))
    aln.append_model(m, atom_files=pdb, align_codes=pdb+chain)
aln.malign()
aln.malign3d()
aln.compare_structures()
aln.id_table(matrix_file='family.mat')
env.dendrogram(matrix_file='family.mat', cluster_cut=-1.0)
```

```python
from modeller import *
from modeller.scripts import complete_pdb

log.verbose()    # request verbose output
env = Environ()
env.libs.topology.read(file='$(LIB)/top_heav.lib') # read topology
env.libs.parameters.read(file='$(LIB)/par.lib') # read parameters

# read model file
mdl = complete_pdb(env, 'TvLDH.B99990002.pdb')

# Assess with DOPE:
s = Selection(mdl)   # all atom selection
s.assess_dope(output='ENERGY_PROFILE NO_REPORT', file='TvLDH.profile',
              normalize_profile=True, smoothing_window=15)
```

```python
import modeller
import matplotlib.pyplot as plt

def r_enumerate(seq):
    """Enumerate a sequence in reverse order"""
    num = len(seq) - 1
    while num >= 0:
        yield num, seq[num]
        num -= 1

def get_profile(profile_file, seq):
    """Read `profile_file` into a Python array, and add gaps corresponding to the alignment sequence `seq`."""
    f = open(profile_file)
    vals = []
    for line in f:
        if not line.startswith('#') and len(line) > 10:
            spl = line.split()
            vals.append(float(spl[-1]))
    
    for n, res in r_enumerate(seq.residues):
        for gap in range(res.get_leading_gaps()):
            vals.insert(n, None)
    
    vals.insert(0, None)
    return vals

e = modeller.Environ()
a = modeller.Alignment(e, file='TvLDH-1bdmA.ali')
template = get_profile('1bdmA.profile', a['1bdmA'])
model = get_profile('TvLDH.profile', a['TvLDH'])

# Plot the template and model profiles in the same plot for comparison
plt.figure(figsize=(10,6))
plt.xlabel('Alignment position')
plt.ylabel('DOPE per-residue score')
plt.plot(model, color='red', linewidth=2, label='Model')
plt.plot(template, color='green', linewidth=2, label='Template')
plt.legend()
plt.savefig('dope_profile.png', dpi=65)
plt.show()
```
